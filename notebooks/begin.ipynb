```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Data import
df = pd.read_csv('./Superwood/Superwood-ID1-22 Jul, 2025.csv')
# Configuration Options
StrainCol = "Strain (%)"
StressCol = "Stress (Pa)"
StrainLim = [1,1.8] #P

# Isolate the linear region (e.g., between 0.1% and 0.5% strain)
linear_region = df[(df[StrainCol] > StrainLim[0]) & (df[StrainCol] < StrainLim[1])]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(linear_region[StrainCol], linear_region[StressCol])

youngs_modulus = slope
print(f'Calculated Young\'s Modulus: {youngs_modulus:.2f} MPa')
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df[StrainCol], df[StressCol], label='Full Stress-Strain Curve')
plt.plot(linear_region[StrainCol], intercept + slope * linear_region[StrainCol], 'r--', linewidth=2, label=f'Linear Fit (E = {youngs_modulus:.2f} MPa)')

plt.xlabel('Strain (%)')
plt.ylabel('Stress (MPa)')
plt.title('Stress-Strain Curve for Sample A')
plt.legend()
plt.grid(True)
plt.show()

# You can save the figure to your project folder
# plt.savefig('Sample_A_Plot.png', dpi=300)
```
